# **Exploring Colombia's rich biodiversity: A closer look at the biomolecule resveratrol**

# **Introduction**

Resveratrol is a biomolecule belonging to the polyphenol family and is widely distributed in various plants, including grapes, peanuts, blackberries, and blueberries [1,2]. Its chemical structure was characterized in 1940 by Takaoka, who successfully isolated it from the roots of Veratrum grandiflorum [1]. This molecule has gained importance in the fields of biochemistry and medicine due to its potential health benefits, which include anti-angiogenic, immunomodulatory, antimicrobial, neurological, antidiabetic, and cardiovascular disease prevention properties [1,2,4,5].

Scientific research has revealed a wide range of potential health benefits of resveratrol. This molecule helps reduce inflammation along with oxidative stress by neutralizing free radicals, thereby contributing to the extended lifespan of various organisms [3]. Other studies have shown improved blood glucose levels and insulin resistance in diabetes [3,7], as well as the prevention of obesity [3,8]. Its efficacy has also been observed in various processes such as mitochondrial dysfunction, inflammation, and apoptosis [3].

The variety of ecosystems in Colombia classify it as a country rich in plant diversity. This attribute allows for the discovery of different species with valuable metabolites such as resveratrol. Among the plants present in Colombia that are notable for their resveratrol content are *Erythrina edulis*, *Vitis vinifera*, *Carica papaya*, *Cissus sicyoides*, among others [9, 10, 11]. These species represent notable examples of the plant richness the country can offer and provide the opportunity to study how resveratrol is produced in different environments and under various environmental conditions.

# **Chemical structure**

Resveratrol (3,5,4'-trihydroxystilbene) is a natural polyphenol with a stilbene structure. Its structure is characterized by the connection of two phenolic rings via a styrene double bond, forming 3,5,4'-trihydroxystilbene (molecular weight: 228.25 g/mol) [1, 12]. This double bond is responsible for the cis and trans isomeric forms of resveratrol. The trans isomer is the more stable form from a steric perspective [1, 12].

In the biomolecule resveratrol, a multitude of synthetic or natural analogs, as well as isomers, derivatives, and conjugates, are known [1]. Additionally, in nature, it is common to find resveratrol bound to simple sugars, forming glucosides.

One of the capabilities of the RDKit tool is its ability to compute molecular descriptors of biomolecules. Molecular descriptors are numerical values that provide quantitative information about various properties of a molecule. Consequently, these descriptors allow for the understanding and comparison of the structural and chemical characteristics of different molecules of interest.

The most pertinent molecular descriptors are enumerated below:

1. **Molecular Weight:** The total weight of the molecule, useful for estimating its size and solubility.

2. **Octanol-Water Partition Coefficient (LogP):** An indicator of a molecule's hydrophobicity.

3. **Number of Carbon Atoms:** The total number of carbon atoms present in the molecule.

4. **Number of Rings:** The number of rings present in the molecule's structural configuration.

5. **Connectivity Index:** A measure of the complexity of the network of bonds in the molecule.

6. **Valence:** The sum of the valence degrees of all the atoms in the molecule.

7. **Hydrogen Bond Donors:** Molecules that donate hydrogen bonds.

8. **Hydrogen Bond Acceptors:** Molecules that accept the donation of electrons.

9. **Melting Temperature:** The temperature at which the molecule undergoes a phase transition from a solid to a liquid state.

```python
# It is possible to continue utilising the tools provided by the RDKit library.
# Calculate different molecular descriptors

smiles = 'Oc1ccc(/C=C/c2cc(O)cc(O)c2)cc1'
mol = Chem.MolFromSmiles(smiles)

mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
hbd = Descriptors.NumHDonors(mol)
hba = Descriptors.NumHAcceptors(mol)
numCarbonAtoms = Descriptors.HeavyAtomCount(mol)
numAromaticRings = Descriptors.NumAromaticRings(mol)
meltingPoint = Descriptors.MolMR(mol)
num_atoms = Descriptors.HeavyAtomCount(mol)
connectivity_index = Descriptors.FractionCSP3(mol)
valency = Descriptors.NumValenceElectrons(mol)

print("Peso molecular:",mw)
print("LogP:", logp)
print("Donantes de enlaces de hidrógeno:", hbd)
print("Aceptores de enlaces de hidrógeno:", hba)
print("Número de atómos de carbono:", numCarbonAtoms)
print("Número de anillos aromaticos:", numAromaticRings)
print("Temperatura de fusión:", meltingPoint)
print("índice de conectividad:", connectivity_index)
print("valencia:", valency)
```

# **Receptor quinone reductase 2 (QR2)**

The protein quinone reductase 2 (QR2) is a key enzyme that plays an essential role in cellular metabolism and protection against oxidative stress. QR2 is a member of the quinone reductase family and is involved in the reduction of quinones and other reactive oxygen species in the body. This protein is widely distributed across various human tissues, and its activity has been observed to be related to the detoxification of toxic compounds and the regulation of cellular redox status. Given its role in cellular metabolism and defense against oxidative damage, QR2 has garnered significant interest in the field of biomedical research and the search for natural compounds that can modulate its activity [14].

Resveratrol, a polyphenol present in various plants, has been identified as a potential ligand for the QR2 protein. Experimental studies have shown that resveratrol binds to QR2 at the enzyme's active site, resulting in the inhibition of its catalytic activity. This interaction between resveratrol and QR2 can have significant implications for human health. Inhibition of QR2 by resveratrol may lead to a reduction in the body's ability to neutralize reactive oxygen species and quinones, which in turn can affect cellular redox regulation and the oxidative-reductive balance. Furthermore, the inhibition of QR2 by resveratrol has garnered interest in the field of oncology, as it has been proposed that this interaction could contribute to the selective apoptosis of cancer cells and have therapeutic potential against certain types of cancer [15,16].

# **Protein structure**

**Primary structure:** The primary structure of a protein is defined as the linear sequence of amino acids that compose it. This sequence is represented by uppercase letters, each corresponding to a specific amino acid in the protein. The primary structure is essential as it determines how the protein folds into its three-dimensional shape and how this influences its biological function [17].

The primary structure of QR2 consists of approximately 205 amino acids and plays a crucial role in its three-dimensional folding and the formation of functional active sites that enable its enzymatic activity. Determining the primary structure of proteins like QR2 is of great importance, as it provides a fundamental basis for understanding their biological functions and interactions with other molecules in the organism.

An example of the primary structure of QR2 is shown below.

```python
# The following line is the usual command for downloading python tools and modules.
!pip install biopython

# Import of the Bio.pdb module that allows us to manipulate PDB files.
from Bio import PDB
# Protein loading
parser = PDB.PDBParser()
structure = parser.get_structure('1SG0', '1SG0.pdb')

# Obtain the primary structure of chains A and B.
chains = ["A", "B"]

for chain in chains:
    # Obtain the amino acid chain.
    aminoacid_chain = structure[0][chain]

    primary_structure = []
    for residue in aminoacid_chain.get_residues():
        if residue.get_resname() not in ["HOH", "WAT"]:
            primary_structure.append(residue.get_resname())

    # Show the primary structure.
    print("Primary chain structure", chain + ":", "".join(primary_structure))
```

**Secondary structure:** It involves the spatial organization of amino acid sequences. This structure can be summarized in two distinct types of organization: beta-pleated sheets (parallel or antiparallel) and alpha helices.

In a beta-pleated sheet, two different regions of a polypeptide chain are positioned side by side and are connected by hydrogen bonds. If two beta sheets run in the same direction, it is a parallel beta-pleated sheet; if they run in opposite directions, it is an antiparallel beta-pleated sheet [18,19]. On the other hand, alpha helices are characterized by the polypeptide chain coiling into a helical shape, held together by hydrogen bonds between every fourth amino acid [18,19]. These two structures are the most common in nature, but it is important to note that they are not the only ones that can occur in polypeptides and proteins.

In QR2, regions can be found that adopt both alpha helices and beta sheets. This structure is essential for the protein's stability and biological function. The combination of these two is indispensable for forming structural domains that perform various functions such as substrate binding, interaction with other proteins, and overall stabilization of the protein.

The following is a visualization of the alpha helices and beta sheets of QR2.

**Tertiary structure:** The tertiary structure is the next level of complexity in protein folding. It corresponds to the three-dimensional arrangement of the entire polypeptide chain, including the alpha helices, beta sheets, and other secondary structures present. This level of folding is determined by the interaction of non-covalent bonds such as electrostatic interactions, Van der Waals forces, and hydrophobic interactions. These interactions enable the protein to acquire its three-dimensional shape [17, 19].

What are these interactions?

* **Electrostatic interactions:** These are attractive or repulsive forces between electrically charged particles. In the tertiary structure of the protein, these interactions can arise between oppositely charged amino acids.

* **Van der Waals forces:** These relatively weak interactions occur between molecules and atoms due to fluctuations in electron densities. Van der Waals forces contribute to the stabilization of three-dimensional protein conformations.

* **Hydrophobic interactions:** These interactions occur between non-polar groups within the protein structure in the presence of water or other polar solvents. The packing of hydrophobic regions is critically important for the stability of the protein's tertiary structure.

We can represent our structure in a better way by following these steps.

```python
# In order to gain access to the contents of the file, it is necessary to open it in read mode.

with open("1SG0.pdb", "r") as pdb_file:
    pdb_data = pdb_file.read()

ligand_lines = []

for line in pdb_data.split("\n"):
    if line.startswith("HETATM"):
      ligand_name = line[17:20].strip()
      if ligand_name == "STL":
        ligand_lines.append(line)

# Convert lines to PDB format.
ligand_pdb = "\n".join(ligand_lines)

# Read the PDB file.
pdb_data = ligand_pdb

# Create an RDKit molecule from PDB file.
mol = Chem.MolFromPDBBlock(pdb_data)

# Guardar la molécula en formato SDF
sdf_supplier = Chem.SDMolSupplier("resveratrol.sdf")  # You can use any existing SDF file as a template.
sdf_writer = Chem.SDWriter("compound_resveratrol.sdf")  # Output file in SDF format, created to write our new molecule.

for template_mol in sdf_supplier:
    new_mol = AllChem.AssignBondOrdersFromTemplate(template_mol, mol)
    sdf_writer.write(new_mol)

sdf_writer.close()

# Analyser of our PDB.
parser = PDBParser()
io=PDBIO()
structure = parser.get_structure('X', '1SG0.pdb')

# A way to look at each string in our PDB and separate them into different files.
for chain in structure.get_chains():
    io.set_structure(chain)
    io.save("1SG0_" + chain.get_id() + ".pdb")

# We assign our py3Dmol method as a view.
view=py3Dmol.view()
# The following lines are used to add the addModel class.
# This class allows us to read PDB files from our chains.
view.addModel(open('1SG0.pdb', 'r').read(),'pdb')
# Here we first read our QR2 file together with resveratrol.
view.addModel(open('compound_resveratrol.sdf', 'r').read(),'sdf')
# Enlargement of all displayed structures.
view.zoomTo()
# Background colour.
view.setBackgroundColor('white')
# Display style of our A, and B chain.
view.setStyle({"stick": {}})
# view.setStyle({'ligand_pdb'},{'stick': {'color':'red'}})
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
view.setStyle({'chain':'B'},{'cartoon': {'color':'green'}})
# Structure display.
view.show()
```
