# **Introducción**

La apigenina (AP) denominada 4’, 5, 7,-trihidroxiflavona, pertenece a un grupo de compuestos llamados flavonoides que se encuentran ampliamente distribuidos en el reino vegetal. Los flavonoides pertenecen al grupo de compuestos polifenólicos que se distribuyen ampliamente en los alimentos de origen vegetal y se dividen en subclases, incluyendo flavonas, flavanonas, flavanoles, isoflavonas y antocianinas [1]. Esto compuestos desempeñan un rol importante de las células vegetales contra microorganismos, insectos e irradiación UV [2]. La evidencia de cultivos celulares, estudios de población animal y humana han sugerido que los flavonoides son beneficiosos para la salud humana y animal. La abundancia de estos compuestos en diferentes alimentos les permite ser nutrientes comunes que son antioxidantes, reguladores estrogénicos y agentes antimicrobianos [1,3]. El flavonoide apigenina desempeña un papel importante en el cuerpo humano como antiinflamatorio, antitóxico y anticancerígeno [3,4].

En la naturaleza, la apigenina está abundantemente presente en frutas comunes como en toronjas, bebidas derivadas de plantas y verduras como perejil, cebollas, naranjas, té, manzanilla, brotes de trigo y en algunos condimentos [3]. En la diversidad colombiana este flavonoide se encuentra presente en diferentes plantas comestibles, como el perejil, la manzanilla, naranjas, guascas y la cebolla. También, Colombia se destaca por su diversidad en plantas y la presencia de la apigenina se puede observar en *Chamaemelum nobile*, *Aloysa citridora*, *Petroselinum crispum*, entre otras [5,6]. La presencia de está biomolécula en la diversidad de plantas es una ventaja que permite generar diferentes investigaciones en Colombia que ayuden a la salud humana.

#**Estructura química**

La apigenina, químicamente conocida como 4',5,7,-trihidroxiflavona, forma parte de la clase de los flavonoides, específicamente de las flavonas. Tiene una fórmula molecular de C15H10O5 y un peso molecular de 270.24 g/mol [1]. Esta molécula biológica se caracteriza por contar con tres grupos hidroxilo como sustituyentes. La eliminación de uno de estos grupos resulta en la estructura fundamental de las flavonas, que se forma a través de selecciones específicas de sustitución de hidroxilos en las posiciones 4’, 5 y 7 del esqueleto estructural.

Una de las funcionalidades que nos aporta la herramienta RDKit es su capacidad para calcular descriptores moleculares de las biomoléculas. Los descriptores moleculares son valores numéricos que proporcionan información cuantitativa sobre diversas propiedades de una molécula. Poe lo anterior estos descriptores nos permiten comprender y comparar las características estructurales y químicas de diferentes moléculas de interés.

A continuación, se nombran los descriptores moleculares de mayor relevancia:

1. Peso molecular: El peso total de la molécula, que es útil para estimar su tamaño y solubilidad.

2. LogP: El coeficiente de partición octanol-agua, que es un indicador de la hidrofobicidad de la molécula.

3. Número de átomos de carbono: El conteo de átomos de carbono en la molécula.

4. Número de anillos: El número de anillos en la estructura de la molécula

5. índice de conectividad: Una medidad de la complejidad de la red de enlaces en la molécula

6. Valencia: La suma de los grados de valencia de todos los átomos de la molécula.

7. Donores: Donantes de enlaces de hidrógeno

8. Aceptores: Aceptores de enlace de hidrógeno

9. Temperatura de fusión: Temperatura a la cual la molécula pasa de estado sólido a estado líquido.

```python
# Podemos seguir aplicando herramientas de la librería RDKit
# Calcular diferentes descriptores moleculares

smiles = 'O=c1cc(-c2ccc(O)cc2)oc2cc(O)cc(O)c12'
mol = Chem.MolFromSmiles(smiles)

mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
hbd = Descriptors.NumHDonors(mol)
hba = Descriptors.NumHAcceptors(mol)
numCarbonAtoms = Descriptors.HeavyAtomCount(mol)
numAromaticRings = Descriptors.NumAromaticRings(mol)
meltingPoint = Descriptors.MolMR(mol)
num_atoms = Descriptors.HeavyAtomCount(mol)
connectivity_index = Descriptors.FractionCSP3(mol)
valency = Descriptors.NumValenceElectrons(mol)

print("Peso molecular:",mw)
print("LogP:", logp)
print("Donantes de enlaces de hidrógeno:", hbd)
print("Aceptores de enlaces de hidrógeno:", hba)
print("Número de atómos de carbono:", numCarbonAtoms)
print("Número de anillos aromaticos:", numAromaticRings)
print("Temperatura de fusión:", meltingPoint)
print("índice de conectividad:", connectivity_index)
print("valencia:", valency)
```

# **Estructura de la proteína**

**Estructura Primaria:** La estructura primaria de una proteína se define como la secuencia lineal de aminoácidos que la componen. Esta secuencia se representa mediante letras mayúsculas, cada una correspondiente a un aminoácido específico en la proteína. La estructura primaria es esencial ya que determina cómo la proteína se pliega tridimensionalmente y cómo esto influye en su función biológica [14]

La estructura primaria de la tanquirasa 2 consta aproximadamente de 1166 aminoácidos y juega un papel crucial en su plegamiento tridimensional y en la formación de los sitios activos funcionales que permiten su actividad enzimática. La determinación de la estructura primaria de proteínas como la TNKS2 es de gran importancia, ya que proporciona una base fundamental para comprender sus funciones biológicas y sus interacciones con otras moléculas en el organismo.

```python
# La siguiente línea es el comando hábitual de descarga de herramientas y modulos de python
!pip install biopython

# Importanción del modulo Bio.pdb que nos permite manipular archivos .PDB
from Bio import PDB
#Cargar la proteína
parser = PDB.PDBParser()
estructura = parser.get_structure('4HKK', '4HKK.pdb')


# Obtener la estructura primaria de las cadenas A y B
cadenas = ["A", "B", "C", "D"]

for cadena in cadenas:
    # Obtener la cadena de aminoácidos
    cadena_aminoacidos = estructura[0][cadena]

    estructura_primaria = []
    for residuo in cadena_aminoacidos.get_residues():
        if residuo.get_resname() not in ["HOH", "WAT"]:
            estructura_primaria.append(residuo.get_resname())

    # Mostrar la estructura primaria
    print("Estructura primaria de la cadena", cadena + ":", "".join(estructura_primaria))
```

**Estructura secundaria:** involucra la organización espacial de las secuencias de aminoácidos. Esta estructura se puede resumir en dos maneras distintias de organización: hoja o lámina plegada beta (paralela o antiparalela) y hélices alfa. En la hoja plegada beta, dos regiones diferentes de una cadena polipéptida se encuentran una al lado de la otra y estan unidas por enlaces de hidrógeno. Si dos láminas beta corren en la misma dirección, entonces es una lámina plegada beta paralela, si tienen direcciones opuestas, será una hoja plegada beta antiparalela [14,15] Por otro lado, las hélices alfa se caracteriza porque la cadena de polipéptidos se enrolla en forma helicoidal, y se mantiene unida por puentes de hidrógeno entre cada cuatro aminoácido [14,15].Estas dos estructuras son las más comunes en la naturaleza pero es pertinente decir que no son las únicas que se pueden presentar en los polipéptidos y proteínas.

En la TNKS2 se pueden encontrar regiones que adoptan tanto hélices alfa como láminas beta. Esta estructura es esencial para la estabilidad y función biológica de la proteína. La combinación de estás dos es indispensable para formar dominios estructurales que desempeñan diferentes funciones cómo: unión a sustratos, interacción con otras proteínas y estabilización global de la proteína.

**Estructura terciaria:** es el siguiente nivel de complejidad en el plegamiento de proteínas. La estructura terciaria de una proteína correspoonde a la disposición tridimensional de la cadena polipeptídica completa, incluyendo las hélices alfa, laminas betas y estructuras secundaria presente. Este nivel de plegamiento está determinado por la interacción de enlaces no covalentes como: interacciones electrostáticas, fuerzas de Van der Waals y interacciones hidrofóbicas. Las interacciones permiten que la proteína adquiera su forma tridimensional [14,16].

¿Qué son estás interacciones?

**Interacciones electrostaticas:** fuerzas atractivas o repulsivas entre partículas cargadas electricamente.

**Fuerzas de Van der Waals:** interacciones debiles que se producen entre moléculas y átomos.

**Interacciones hidrofóbicas:** interacciones entre moléculas no polares y se producen en presencia de agua o otros solventes polares.

Las interacciones están presentes en el acoplamiento de la apigenina con la TNKS2. A continuación se presenta una visualización de los residuos implicados y las interacciones presentes en el acoplamiento. También la representación en 3D del ligando (apigenina)con su receptor TNKS2.

Figura 3. Interacción de la apigenina con TNKS2.

```python
# Abrimos el archivo en modo lectura

with open("4HKK.pdb", "r") as pdb_file:
    pdb_data = pdb_file.read()

# Buscar y extraer las líneas correspondientes al ligando
ligand_lines = []

for line in pdb_data.split("\n"):
    if line.startswith("HETATM"):
      ligand_name = line[17:20].strip()
      if ligand_name == "AGI":
        ligand_lines.append(line)

# Convertir las líneas en formato PDB
ligand_pdb = "\n".join(ligand_lines)

# Leer el archivo PDB
pdb_data = ligand_pdb

# Crear una molécula RDKit desde el archivo PDB
mol = Chem.MolFromPDBBlock(pdb_data)

# Guardar la molécula en formato SDF
sdf_supplier = Chem.SDMolSupplier("apigenina.sdf")  # Puedes usar cualquier archivo SDF existente como plantilla
sdf_writer = Chem.SDWriter("compuesto_apigenina.sdf")  # Archivo de salida en formato SDF, se crea para escribir nuestra nueva molécula

for template_mol in sdf_supplier:
    new_mol = AllChem.AssignBondOrdersFromTemplate(template_mol, mol)
    sdf_writer.write(new_mol)

sdf_writer.close()

#analizador de nuestro .PDB
parser = PDBParser()
io=PDBIO()
structure = parser.get_structure('X', '4HKK.pdb')

# Una forma de observar cada cadena de nuestro PDB y separar las en archivos diferentes
for chain in structure.get_chains():
    io.set_structure(chain)
    io.save("4HKK_" + chain.get_id() + ".pdb")

# Asignamos nuestro método py3Dmol como una vista
view=py3Dmol.view()

# Las siguiente líneas se utiliza para agregar la clase addModel

# Esta clase nos permite leer los archivos PDB de nuestras cadenas
view.addModel(open('4HKK.pdb', 'r').read(),'pdb')
#Aquí leemos primero nuestro archivo de DAPK1 junto Kaempferol.
view.addModel(open('compuesto_apigenina.sdf', 'r').read(),'sdf')
# Ampliación de todas las estructuras visualizadas
view.zoomTo()
# Color de fondo
view.setBackgroundColor('white')
# Estilo de visualización de nuestra cadena A, B, C y D
view.setStyle({"stick": {}})
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
view.setStyle({'chain':'B'},{'cartoon': {'color':'green'}})
view.setStyle({'chain':'C'},{'cartoon': {'color':'green'}})
view.setStyle({'chain':'D'},{'cartoon': {'color':'yellow'}})
# Visualización de estructura
view.show()
```
