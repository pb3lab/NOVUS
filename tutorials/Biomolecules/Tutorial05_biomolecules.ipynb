# **Exploring Colombia's rich biodiversity: A closer look at the biomolecule luteolin**

# **Introduction**

Luteolin (LU), also known as 3',4',5,7-tetrahydroxyflavone, belongs to a group of natural compounds called flavonoids, which are widely distributed in the plant kingdom [1, 7]. Flavonoids are polyphenols that play an important role in defending plant cells against microorganisms, insects, and UV irradiation [1, 2]. Evidence from cell culture, animal, and human population studies suggests that flavonoids are also beneficial to human and animal health [3]. The abundance of these compounds in different foods makes them common nutrients, serving as antioxidants, estrogen regulators, and antimicrobial agents [1]. Specifically, the flavonoid luteolin plays a significant role in the human body as an antioxidant, free radical scavenger, anti-inflammatory agent, and immune system modulator [4]. Additionally, it is an active compound against several types of cancer, including colon, pancreatic, lung, and kidney cancer [4, 5].

In Colombia's diverse flora, luteolin is found in various edible plants, such as pepper, celery, carrots, and spinach [6]. It is also present in vegetables and fruits like broccoli, onion leaves, parsley, sweet peppers, and chrysanthemum flowers [7]. This biomolecule is found worldwide in species such as *Sedum sarmentosum*, *Vicia sylvatica*, *Vicia americana*, and *Passiflora incarnata*, among others [4]. Colombia, known for its plant diversity, also features luteolin in species like *Cymbopogon citratus*, *Cassia grandis*, *Passiflora ligularis*, and *Theobroma cacao* [4]. The widespread presence of this flavonoid in various plants is advantageous, as it enables extensive research in Colombia to support human health.

# **Chemical structure**

Luteolin has a C6-C3-C6 structure and possesses two benzene rings (A, B), a third oxygen-containing ring (C), and a 2-3 carbon double bond [1, 4]. Additionally, luteolin has hydroxyl groups at carbon positions 5, 7, 3', and 4' [1, 8]. The hydroxyl residues and the 2-3 carbon double bond are important structural features of this biomolecule, which are associated with its biochemical and biological activities [1, 9].

One of the capabilities of the RDKit tool is its ability to compute molecular descriptors of biomolecules. Molecular descriptors are numerical values that provide quantitative information about various properties of a molecule. Consequently, these descriptors allow for the understanding and comparison of the structural and chemical characteristics of different molecules of interest.

The most pertinent molecular descriptors are enumerated below:

1. **Molecular Weight:** The total weight of the molecule, useful for estimating its size and solubility.

2. **Octanol-Water Partition Coefficient (LogP):** An indicator of a molecule's hydrophobicity.

3. **Number of Carbon Atoms:** The total number of carbon atoms present in the molecule.

4. **Number of Rings:** The number of rings present in the molecule's structural configuration.

5. **Connectivity Index:** A measure of the complexity of the network of bonds in the molecule.

6. **Valence:** The sum of the valence degrees of all the atoms in the molecule.

7. **Hydrogen Bond Donors:** Molecules that donate hydrogen bonds.

8. **Hydrogen Bond Acceptors:** Molecules that accept the donation of electrons.

9. **Melting Temperature:** The temperature at which the molecule undergoes a phase transition from a solid to a liquid state.

```python
# It is possible to continue utilising the tools provided by the RDKit library.
# Calculate different molecular descriptors

smiles = 'O=c1cc(-c2ccc(O)c(O)c2)oc2cc(O)cc(O)c12'
mol = Chem.MolFromSmiles(smiles)

mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
hbd = Descriptors.NumHDonors(mol)
hba = Descriptors.NumHAcceptors(mol)
numCarbonAtoms = Descriptors.HeavyAtomCount(mol)
numAromaticRings = Descriptors.NumAromaticRings(mol)
meltingPoint = Descriptors.MolMR(mol)
num_atoms = Descriptors.HeavyAtomCount(mol)
connectivity_index = Descriptors.FractionCSP3(mol)
valency = Descriptors.NumValenceElectrons(mol)

print("Peso molecular:",mw)
print("LogP:", logp)
print("Donantes de enlaces de hidrógeno:", hbd)
print("Aceptores de enlaces de hidrógeno:", hba)
print("Número de atómos de carbono:", numCarbonAtoms)
print("Número de anillos aromaticos:", numAromaticRings)
print("Temperatura de fusión:", meltingPoint)
print("índice de conectividad:", connectivity_index)
print("valencia:", valency)
```

# **Receptor tankyrase 2**

Tanquirases are members of the diphtheria toxin-like ADP ribosyltransferases (ARTD) family [10,11]. Tanquirases 1 and 2 are multidomain proteins consisting of four domains: a histidine-proline-serine-rich domain, ankyrin repeats, a sterile alpha motif (SAM), and a catalytic ART domain. These proteins are located in mitotic centrosomes, the Golgi apparatus, the cytoplasm, and nuclear pore complexes [10]. Tanquirases are attractive drug targets for cancer due to their demonstrated anticancer therapeutic potential, which is related to their roles in telomere homeostasis, mitosis, and Wnt signaling [10,12].

Research has shown that luteolin is a less potent inhibitor of tanquirase compared to apigenin [13]. Luteolin has demonstrated selectivity towards tanquirase 2, and it is suggested that some flavonoids (flavones) could be highly selective inhibitors of tanquirase [10]. This interaction may lead to beneficial effects on human health, as flavonoids are known to have antioxidant, antiviral, anti-inflammatory, neuroprotective, and anticancer properties [1,5,7].

# **Protein structure**

**Primary structure:** The primary structure of a protein is defined as the linear sequence of amino acids that compose it. This sequence is represented by uppercase letters, each corresponding to a specific amino acid in the protein. The primary structure is essential as it determines how the protein folds into its three-dimensional shape and how this influences its biological function [14].

The primary structure of tankyrase 2 consists of approximately 1166 amino acids and plays a crucial role in its three-dimensional folding and the formation of functional active sites that enable its enzymatic activity. Determining the primary structure of proteins like TNKS2 is of great importance, as it provides a fundamental basis for understanding their biological functions and interactions with other molecules in the organism.

An example of the primary structure of tankyrase 2 is shown below.

```python
# The following line is the usual command for downloading python tools and modules.
!pip install biopython

# Import of the Bio.pdb module that allows us to manipulate PDB files.
from Bio import PDB
# Protein loading
parser = PDB.PDBParser()
structure = parser.get_structure('4HKN', '4HKN.pdb')

# Obtain the primary structure of chains A and B.
chains = ["A", "C"]

for chain in chains:
    # Obtain the amino acid chain.
    aminoacid_chain = structure[0][chain]

    primary_structure = []
    for residue in aminoacid_chain.get_residues():
        if residue.get_resname() not in ["HOH", "WAT"]:
            primary_structure.append(residue.get_resname())

    # Show the primary structure.
    print("Primary chain structure", chain + ":", "".join(primary_structure))
```

**Secondary structure:** It involves the spatial organization of amino acid sequences. This structure can be summarized in two distinct types of organization: beta-pleated sheets (parallel or antiparallel) and alpha helices. In the beta-pleated sheet, two different regions of a polypeptide chain lie side by side and are held together by hydrogen bonds. If two beta sheets run in the same direction, it is a parallel beta-pleated sheet; if they run in opposite directions, it is an antiparallel beta-pleated sheet [14,15]. On the other hand, alpha helices are characterized by the polypeptide chain coiling into a helical shape, held together by hydrogen bonds between every fourth amino acid [14,15]. These two structures are the most common in nature, but it is important to note that they are not the only ones that can occur in polypeptides and proteins.

In TNKS2, regions that adopt both alpha helices and beta sheets can be found. This structure is essential for the protein's stability and biological function. The combination of these two elements is indispensable for forming structural domains that perform various functions such as substrate binding, interaction with other proteins, and overall stabilization of the protein.

The following is a visualization of the alpha helices and beta sheets of tankyrase 2.

**Tertiary structure:** The tertiary structure is the next level of complexity in protein folding. It corresponds to the three-dimensional arrangement of the entire polypeptide chain, including the alpha helices, beta sheets, and other secondary structures present. This level of folding is determined by the interaction of non-covalent bonds such as electrostatic interactions, Van der Waals forces, and hydrophobic interactions. These interactions enable the protein to acquire its three-dimensional shape [14,16].

What are these interactions?

* **Electrostatic interactions:** These are attractive or repulsive forces between electrically charged particles. In the tertiary structure of the protein, these interactions can arise between oppositely charged amino acids.

* **Van der Waals forces:** These relatively weak interactions occur between molecules and atoms due to fluctuations in electron densities. Van der Waals forces contribute to the stabilization of three-dimensional protein conformations.

* **Hydrophobic interactions:** These interactions occur between non-polar groups within the protein structure in the presence of water or other polar solvents. The packing of hydrophobic regions is critically important for the stability of the protein's tertiary structure.

The aforementioned interactions are crucial for the docking of luteolin with TNKS2, allowing the molecule to bind specifically and affect the enzyme. We can represent our structure in a better way by following these steps.

```python
# In order to gain access to the contents of the file, it is necessary to open it in read mode.

with open("4HKN.pdb", "r") as pdb_file:
    pdb_data = pdb_file.read()

ligand_lines = []

for line in pdb_data.split("\n"):
    if line.startswith("HETATM"):
      ligand_name = line[17:20].strip()
      if ligand_name == "LU2":
        ligand_lines.append(line)

# Convert lines to PDB format.
ligand_pdb = "\n".join(ligand_lines)

# Read the PDB file.
pdb_data = ligand_pdb

# Create an RDKit molecule from PDB file.
mol = Chem.MolFromPDBBlock(pdb_data)

# Guardar la molécula en formato SDF
sdf_supplier = Chem.SDMolSupplier("luteolin.sdf")  # You can use any existing SDF file as a template.
sdf_writer = Chem.SDWriter("compound_luteolin.sdf")  # Output file in SDF format, created to write our new molecule.

for template_mol in sdf_supplier:
    new_mol = AllChem.AssignBondOrdersFromTemplate(template_mol, mol)
    sdf_writer.write(new_mol)

sdf_writer.close()

# Analyser of our PDB.
parser = PDBParser()
io=PDBIO()
structure = parser.get_structure('X', '4HKN.pdb')

# A way to look at each string in our PDB and separate them into different files.
for chain in structure.get_chains():
    io.set_structure(chain)
    io.save("4HKN_" + chain.get_id() + ".pdb")

# We assign our py3Dmol method as a view.
view=py3Dmol.view()
# The following lines are used to add the addModel class.
# This class allows us to read PDB files from our single A-chain.
view.addModel(open('4HKN.pdb', 'r').read(),'pdb')
# Here we first read our DAPK1 file together with Kaempferol.
view.addModel(open('compound_luteolin.sdf', 'r').read(),'sdf')
# Enlargement of all displayed structures.
view.zoomTo()
# Background colour.
view.setBackgroundColor('white')
# Display style of our A and B chain.
view.setStyle({"stick": {}})
# view.setStyle({'ligand_pdb'},{'stick': {'color':'red'}})
view.setStyle({'chain':'A'},{'cartoon': {'color':'orange'}})
view.setStyle({'chain':'C'},{'cartoon': {'color':'green'}})
# Structure display.
view.show()
```
