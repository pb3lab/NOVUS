# **Exploring Colombia's rich biodiversity: A closer look at the biomolecule quercetin**

# **Introduction**

Flavonoids are a group of secondary metabolites widely distributed in the plant kingdom, characterized by their diphenylpropane structure. These bioactive compounds have been extensively studied due to their beneficial properties for human health and their role in the prevention of chronic diseases [1]. Among the most prominent flavonoids is quercetin, a low molecular weight compound (302.2 g/mol) with a chemical structure characterized by phenol rings and a pyran ring [2].

Quercetin has attracted the attention of the scientific community due to its antioxidant, anti-inflammatory, anticarcinogenic, and cardioprotective properties [3,4]. The antioxidant properties of this biomolecule have been shown to be effective in neutralizing reactive oxygen species, reducing oxidative stress, and protecting cells from free radical damage [5]. Quercetin has also demonstrated anti-inflammatory effects by inhibiting the release of certain pro-inflammatory cytokines and inflammatory enzymes, suggesting its potential in treating chronic inflammatory diseases [6]. Additionally, it has been observed that quercetin can modulate various cell signaling pathways involved in cell survival and proliferation, which could have implications for the treatment of cancer and related diseases [7,8].

The presence of quercetin in Colombia is reflected in a wide variety of endemic plants that contain this biomolecule and other important phytochemical compounds [9]. This biomolecule is found in different plant families within Colombia's biodiversity, such as Asteraceae, Solanaceae, Fabaceae, Passifloraceae, Moraceae, among others [10]. One example is *Passiflora edulis*, which contains quercetin in its fruits. Studying this biomolecule and its distribution in different plant species in Colombia can contribute to the knowledge of the country's biological diversity and aid in identifying new sources of bioactive compounds with therapeutic and health applications [10].

# **Chemical structure**

Quercetin is a flavonoid belonging to the flavonol class, characterized by a tetracyclic structure with fused carbon rings (A, B, and C) along with a pyran ring (D ring) [11]. This arrangement gives it a distinctive shape that contributes to its stability and solubility in both polar and non-polar compounds [12]. Additionally, quercetin contains asymmetric centers, allowing it to exist in multiple isomeric forms. This can result in different pharmacological properties and the ability to interact with specific proteins and receptors in the body [13].

In the structure of quercetin, it is important to mention the presence of hydroxyl groups (-OH) at specific positions on the A and B rings. These groups are essential for its antioxidant activity, its ability to neutralize free radicals, and its role in protecting cells from oxidative damage [14]. Additionally, these hydroxyl groups may be involved in quercetin's interactions with enzymes and cellular receptors, modulating its biological effects and therapeutic potential [7].

One of the capabilities of the RDKit tool is its ability to compute molecular descriptors of biomolecules. Molecular descriptors are numerical values that provide quantitative information about various properties of a molecule. Consequently, these descriptors allow for the understanding and comparison of the structural and chemical characteristics of different molecules of interest.

The most pertinent molecular descriptors are enumerated below:

1. **Molecular Weight:** The total weight of the molecule, useful for estimating its size and solubility.

2. **Octanol-Water Partition Coefficient (LogP):** An indicator of a molecule's hydrophobicity.

3. **Number of Carbon Atoms:** The total number of carbon atoms present in the molecule.

4. **Number of Rings:** The number of rings present in the molecule's structural configuration.

5. **Connectivity Index:** A measure of the complexity of the network of bonds in the molecule.

6. **Valence:** The sum of the valence degrees of all the atoms in the molecule.

7. **Hydrogen Bond Donors:** Molecules that donate hydrogen bonds.

8. **Hydrogen Bond Acceptors:** Molecules that accept the donation of electrons.

9. **Melting Temperature:** The temperature at which the molecule undergoes a phase transition from a solid to a liquid state.

```python
# It is possible to continue utilising the tools provided by the RDKit library.
# Calculate different molecular descriptors

smiles = 'O=c1c(O)c(-c2ccc(O)c(O)c2)oc2cc(O)cc(O)c12'
mol = Chem.MolFromSmiles(smiles)

mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
hbd = Descriptors.NumHDonors(mol)
hba = Descriptors.NumHAcceptors(mol)
numCarbonAtoms = Descriptors.HeavyAtomCount(mol)
numAromaticRings = Descriptors.NumAromaticRings(mol)
meltingPoint = Descriptors.MolMR(mol)
num_atoms = Descriptors.HeavyAtomCount(mol)
connectivity_index = Descriptors.FractionCSP3(mol)
valency = Descriptors.NumValenceElectrons(mol)

print("Peso molecular:",mw)
print("LogP:", logp)
print("Donantes de enlaces de hidrógeno:", hbd)
print("Aceptores de enlaces de hidrógeno:", hba)
print("Número de atómos de carbono:", numCarbonAtoms)
print("Número de anillos aromaticos:", numAromaticRings)
print("Temperatura de fusión:", meltingPoint)
print("índice de conectividad:", connectivity_index)
print("valencia:", valency)
```

# **Protein structure**

**Primary structure:** The primary structure of a protein is defined as the linear sequence of amino acids that compose it. This sequence is represented by capital letters, each corresponding to a specific amino acid in the protein. The primary structure is essential as it determines how the protein folds three-dimensionally and how this influences its biological function [20].

The primary structure of DAPK1 is essential for its biological function, as it determines the three-dimensional shape the protein will take and how this structure will affect its activity and its ability to interact with other molecules in the cell. The amino acid sequence in the primary structure of DAPK1 is encoded by the DAPK1 gene and consists of approximately 1,430 amino acids [2]. Each amino acid has its own chemical and functional properties, which contribute to the final conformation of the protein and its biological activity.

An example of the primary structure of DAPK1 is shown below.

```python
# The following line is the usual command for downloading python tools and modules.
!pip install biopython

# Import of the Bio.pdb module that allows us to manipulate PDB files.
from Bio import PDB
# Protein loading
parser = PDB.PDBParser()
structure = parser.get_structure('5AUW', '5AUW.pdb')

# Obtain the primary structure of chains A and B.
chains = ["A"]

for chain in chains:
    # Obtain the amino acid chain.
    aminoacid_chain = structure[0][chain]

    primary_structure = []
    for residue in aminoacid_chain.get_residues():
        if residue.get_resname() not in ["HOH", "WAT"]:
            primary_structure.append(residue.get_resname())

    # Show the primary structure.
    print("Primary chain structure", chain + ":", "".join(primary_structure))
```

**Secondary structure:** This involves the spatial organization of amino acid sequences. The secondary structure can be summarized in two main types of organization: beta-folded sheets (parallel or antiparallel) and alpha helices.

In the beta-folded sheet, two different regions of a polypeptide chain lie next to each other and are linked by hydrogen bonds. If two beta-sheets run in the same direction, it is a parallel beta-folded sheet; if they run in opposite directions, it is an antiparallel beta-folded sheet [20,21]. On the other hand, alpha helices are characterized by the polypeptide chain being helically coiled and held together by hydrogen bonds between every fourth amino acid [20,21]. These two structures are the most common in nature, but it is important to note that they are not the only ones that can occur in polypeptides and proteins.

In DAPK1, regions can be found that adopt both alpha helices and beta sheets. This structural diversity is essential for the stability and biological function of the protein. The combination of these two elements is indispensable for forming structural domains that perform various functions, such as binding to substrates, interacting with other proteins, and providing global stabilization of the protein.

**Tertiary structure:** The tertiary structure of a protein represents the next level of complexity in protein folding. It corresponds to the three-dimensional arrangement of the entire polypeptide chain, including the alpha helices, beta sheets, and other secondary structures present. This level of folding is determined by interactions involving non-covalent bonds such as electrostatic interactions, van der Waals forces, and hydrophobic interactions. These interactions allow the protein to acquire its specific three-dimensional shape [20,22].

What are these interactions?

* **Electrostatic interactions:** These are attractive or repulsive forces between electrically charged particles. In the tertiary structure of the protein, these interactions can arise between oppositely charged amino acids.

* **Van der Waals forces:** These relatively weak interactions occur between molecules and atoms due to fluctuations in electron densities. Van der Waals forces contribute to the stabilization of three-dimensional protein conformations.

* **Hydrophobic interactions:** These interactions occur between non-polar groups within the protein structure in the presence of water or other polar solvents. The packing of hydrophobic regions is critically important for the stability of the protein's tertiary structure.

```python
# In order to gain access to the contents of the file, it is necessary to open it in read mode.

with open("5AUW.pdb", "r") as pdb_file:
    pdb_data = pdb_file.read()

ligand_lines = []

for line in pdb_data.split("\n"):
    if line.startswith("HETATM"):
      ligand_name = line[17:20].strip()
      if ligand_name == "QUE":
        ligand_lines.append(line)

# Convert lines to PDB format.
ligand_pdb = "\n".join(ligand_lines)

# Read the PDB file.
pdb_data = ligand_pdb

# Create an RDKit molecule from PDB file.
mol = Chem.MolFromPDBBlock(pdb_data)

# Guardar la molécula en formato SDF
sdf_supplier = Chem.SDMolSupplier("quercetine.sdf")  # You can use any existing SDF file as a template.
sdf_writer = Chem.SDWriter("compuesto_quercetine.sdf")  # Output file in SDF format, created to write our new molecule.

for template_mol in sdf_supplier:
    new_mol = AllChem.AssignBondOrdersFromTemplate(template_mol, mol)
    sdf_writer.write(new_mol)

sdf_writer.close()

# Analyser of our PDB.
parser = PDBParser()
io=PDBIO()
structure = parser.get_structure('X', '5AUW.pdb')

# A way to look at each string in our PDB and separate them into different files.
for chain in structure.get_chains():
    io.set_structure(chain)
    io.save("5AUW_" + chain.get_id() + ".pdb")

# We assign our py3Dmol method as a view.
view=py3Dmol.view()
# The following lines are used to add the addModel class.
# This class allows us to read PDB files from our single A-chain.
view.addModel(open('5AUW.pdb', 'r').read(),'pdb')
# Here we first read our DAPK1 file together with Kaempferol.
view.addModel(open('compuesto_quercetine.sdf', 'r').read(),'sdf')
# Enlargement of all displayed structures.
view.zoomTo()
# Background colour.
view.setBackgroundColor('white')
# Display style of our A and B chain.
view.setStyle({"stick": {}})
# view.setStyle({'ligand_pdb'},{'stick': {'color':'red'}})
view.setStyle({'chain':'A'},{'cartoon': {'color':'blue'}})
# Structure display.
view.show()
```
