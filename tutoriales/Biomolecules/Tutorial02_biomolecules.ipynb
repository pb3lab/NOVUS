# **Introducción**

Los flavonoides son un grupo de metabolitos secundarios que se encuentran ampliamente en el reino vegetal y se caracterizan por su estructura de difenilpropano. Estos compuestos se han investigado extensamente debido a su importancia en la reducción del riesgo de enfermedades crónicas y su papel en la protección contra el cáncer y las enfermedades cardiovasculares [1].  

Uno de estos flavonoides es el kaempferol, un compuesto de bajo peso molecular (286,2g/mol) de color amarillo [4]. El kaempferol se encuentra comúnmente en alimentos de origen vegetal y en plantas utilizadas en medicina tradicional. Está biomolécula actúa como un antioxidante que combate el estrés oxidativo y también ha mostrado propiedades antibacterianas [2]. Además, se ha relacionado con una serie de beneficios para la salud humana, incluyendo la protección contra enfermedades neurodegenerativas, efectos antiinflamatorios y propiedades anticancerígenas [3,4]. Asimismo, se ha observado que el kaempferol tiene la capacidad de modular expresión de genes, señalización celular y puede tener un efecto beneficioso sobre el sistema inmunológico [4].

La biomolécula del kaempferol se distribuye ampliamente en diferentes géneros de plantas, como Delphinium, Camellia, Berberis, Citrus, Brassica, Allium, Malus, entre otros. Además, se ha identificado plantas medicinales, tales como *Acacia nilotica, Crocus savitus L, Rosmarinus officinalis L* [1,2]. En Colombia el kaempferol se encuentra en diversas familias de plantas nativas del país. Entre ellas se destacan las familias Asteraceae y Solanaceae, que cuentan con especies notables que contienen esta biomolécula. Un ejemplo relevante es la especie *Espeletia grandiflora*, perteneciente a la familia Asteraceae, la cual se encuentra en las altas montañas de los páramos colombianos y posee altas concentraciones de kaempferol en sus tejidos [5,6]. Otro ejemplo, es la especie *Brugmansia suaveolens*, de la familia Solanaceae, también conocida como "borrachero", es otra planta endémica de Colombia que contiene este compuesto y ha sido objeto de interés científico debido a sus propiedades medicinales y ornamentales. La presencia del kaempferol en estas especies y su distribución en diferentes familias demuestra la diversidad biológica de Colombia y su potencial como fuentes de compuestos con propiedades beneficiosas.

# **Estructura química**

La estructura química del kaempferol se caracteriza por ser un flavonol, perteneciente a la clase de los flavonoides, y está compuesto por tres anillos de carbono fusionados (A, B y C) y un anillo tipo pirano (D) [9]. Esta disposición tetracíclica le otorga su forma distintiva y contribuye a su estabilidad y solubilidad en compuestos polares y no polares [4,9,10]. Además, el kaempferol contiene centros asimétricos que le confieren la capacidad de existir en múltiples formas isoméricas.  Estas disposiciones son variaciones en la orientación espacial de los átomos alrededor de estos centros, lo que puede dar lugar a diferentes propiedades farmacológicas [11].

En la estructura del kaempferol es importante mencionar la presencia de grupos hidroxilos (-OH) en posiciones específicas de los anillos A y B. Estos grupos son fundamentales para su actividad antioxidante y su capacidad de neutralizar radicales libres, así como para proteger las células de daño oxidativo [12]. Además, estos grupos hidroxilos pueden estar involucrados en la interacción de enzimas y receptores celulares, lo que modula sus efectos biológicos [11,12]

Una de las funcionalidades que nos aporta la herramienta RDKit es su capacidad para calcular descriptores moleculares de las biomoléculas. Los descriptores moleculares son valores numéricos que proporcionan información cuantitativa sobre diversas propiedades de una molécula. Estos descriptores nos permiten comprender y comparar las características estructurales y químicas de diferentes moléculas de interés.

A continuación, se nombran los descriptores moleculares de mayor relevancia:

1. Peso molecular: El peso total de la molécula, que es útil para estimar su tamaño y solubilidad.

2. LogP. El coeficiente de partición octanol-agua, que es un indicador de la hidrofobicidad de la molécula.
3. Número de átomos de carbono: El conteo de átomos de carbono en la molécula.

4. Número de anillos: El número de anillos en la estructura de la molécula

5. índice de conectividad: Una medida de la complejidad de la red de enlaces en la molécula

6. Valencia: La suma de los grados de valencia de todos los átomos de la molécula.

7. Donores: Donantes de enlaces de hidrógeno

8. Aceptores: Aceptores de enlace de hidrógeno

9. Temperatura de fusión: Temperatura a la cual la molécula pasa de estado sólido a estado líquido.  


```python
# Podemos seguir aplicando herramientas de la librería RDKit
# Calcular diferentes descriptores moleculares

smiles = 'O=c1c(O)c(-c2ccc(O)cc2)oc2cc(O)cc(O)c12'
mol = Chem.MolFromSmiles(smiles)

mw = Descriptors.MolWt(mol)
logp = Descriptors.MolLogP(mol)
hbd = Descriptors.NumHDonors(mol)
hba = Descriptors.NumHAcceptors(mol)
numCarbonAtoms = Descriptors.HeavyAtomCount(mol)
numAromaticRings = Descriptors.NumAromaticRings(mol)
meltingPoint = Descriptors.MolMR(mol)
num_atoms = Descriptors.HeavyAtomCount(mol)
connectivity_index = Descriptors.FractionCSP3(mol)
valency = Descriptors.NumValenceElectrons(mol)

print("Peso molecular:",mw)
print("LogP:", logp)
print("Donantes de enlaces de hidrógeno:", hbd)
print("Aceptores de enlaces de hidrógeno:", hba)
print("Número de atómos de carbono:", numCarbonAtoms)
print("Número de anillos aromaticos:", numAromaticRings)
print("Temperatura de fusión:", meltingPoint)
print("índice de conectividad:", connectivity_index)
print("valencia:", valency)
```

# **Estructura de la proteína**

**Estructura Primaria**: La estructura primaria de una proteína se define como la secuencia lineal de aminoácidos que la componen. Esta secuencia se representa mediante letras mayúsculas, cada una correspondiente a un aminoácido específico en la proteína. La estructura primaria es esencial ya que determina cómo la proteína se pliega tridimensionalmente y cómo esto influye en su función biológica [18].

La estructura primaria de la DAPK1 es esencial para su función biológica, debido a que determina la forma tridimensional que adoptará la proteína y cómo esta estructura afectará su actividad y su capacidad para interactuar con otras moléculas en la célula. La secuencia de aminoácidos en la estructura primaria de la DAPK1 está codificada por el gen DAPK1 y consta de aproximadamente 1430 aminoácidos. Cada uno de los aminoácidos tiene su propia propiedad química y funcional, lo que contribuye a la conformación final de la proteína y a su actividad biológica.

A continuación, se presenta un ejemplo de la estructura primaria de la DAPK1.

```python
# La siguiente línea es el comando hábitual de descarga de herramientas y modulos de python
!pip install biopython

# Importanción del modulo Bio.pdb que nos permite manipular archivos .PDB
from Bio import PDB
#Cargar la proteína
parser = PDB.PDBParser()
estructura = parser.get_structure('5AUX', '5AUX.pdb')


# Obtener la estructura primaria de las cadenas A
cadenas = ["A"]

for cadena in cadenas:
    # Obtener la cadena de aminoácidos
    cadena_aminoacidos = estructura[0][cadena]

    estructura_primaria = []
    for residuo in cadena_aminoacidos.get_residues():
        if residuo.get_resname() not in ["HOH", "WAT"]:
            estructura_primaria.append(residuo.get_resname())

    # Mostrar la estructura primaria
    print("Estructura primaria de la cadena", cadena + ":", "".join(estructura_primaria))
```

**Estructura secundaria**: Involucra la organización espacial de las secuencias de aminoácidos. Esta estructura se puede resumir en dos maneras distintas de organización: hoja o lámina plegada beta (paralela o antiparalela) y hélices alfa.

En la lámina plegada beta, dos regiones diferentes de una cadena polipéptida se encuentran una al lado de la otra y están unidas por enlaces de hidrógeno. Si dos láminas beta corren en la misma dirección, entonces es una lámina plegada beta paralela, si tienen direcciones opuestas, será una hoja plegada beta antiparalela [4,18, 19]. Por otro lado, las hélices alfa se caracteriza porque la cadena de polipéptidos se enrolla en forma helicoidal, y se mantiene unida por puentes de hidrógeno entre cada cuatro aminoácido [18,19].Estas dos estructuras son las más comunes en la naturaleza pero es pertinente decir que no las únicas que se pueden presentar en los polipéptidos y proteínas.

En DAPK1 se pueden encontrar regiones que adoptan tanto hélices alfa como láminas beta. Esta estructura es esencial para la estabilidad y función biológica de la proteína y la combinación de estás dos es indispensable para formar dominios estructurales que desempeñan diferentes funciones cómo: unión a sustratos, interacción con otras proteínas y estabilización global de la proteína.

**Estructura terciaria**: Es el siguiente nivel de complejidad en el plegamiento de proteínas. La estructura terciaria de una proteína corresponde a la disposición tridimensional de la cadena polipeptídica completa, incluyendo las hélices alfa, laminas betas y estructuras secundaria presente. Este nivel de plegamiento está determinado por la interacción de enlaces no covalentes como: interacciones electrostáticas, fuerzas de van der Waals e interacciones hidrofóbicas. Las interacciones permiten que la proteína adquiera su forma tridimensional [18,20].

¿Qué son estás interacciones?

* Interacciones electrostáticas: fuerzas atractivas o repulsivas entre partículas cargadas eléctricamente.

* Fuerzas de van der Waals: interacciones débiles que se producen entre moléculas y átomos.

* Interacciones hidrofóbicas: interacciones entre moléculas no polares y se producen en presencia de agua u otros solventes polares.


Las interacciones mencionadas son cruciales en el acoplamiento del kaempferol con DAPK1, permitiendo que la molécula se una de manera específica y tenga un efecto sobre la enzima. A continuación, se presenta una visualización de los residuos implicados y las interacciones presentes en el acoplamiento. También la representación en 3D del ligando junto a su receptor.

```python
#Abrimos el archivo en modo lectura

with open("5AUX.pdb", "r") as pdb_file:
    pdb_data = pdb_file.read()

# Buscar y extraer las líneas correspondientes al ligando
ligand_lines = []

for line in pdb_data.split("\n"):
    if line.startswith("HETATM"):
      ligand_name = line[17:20].strip()
      if ligand_name == "KMP":
        ligand_lines.append(line)


# Convertir las líneas en formato PDB
ligand_pdb = "\n".join(ligand_lines)

# Leer el archivo PDB
pdb_data = ligand_pdb

# Crear una molécula RDKit desde el archivo PDB
mol = Chem.MolFromPDBBlock(pdb_data)

# Guardar la molécula en formato SDF
sdf_supplier = Chem.SDMolSupplier("kaempferol.sdf")  # Puedes usar cualquier archivo SDF existente como plantilla
sdf_writer = Chem.SDWriter("compuesto_kaempferol.sdf")  # Archivo de salida en formato SDF, se crea para escribir nuestra nueva molécula

for template_mol in sdf_supplier:
    new_mol = AllChem.AssignBondOrdersFromTemplate(template_mol, mol)
    sdf_writer.write(new_mol)

sdf_writer.close()

#analizador de nuestro .PDB
parser = PDBParser()
io=PDBIO()
structure = parser.get_structure('X', '5AUX.pdb')

# Una forma de observar cada cadena de nuestro PDB y separar las en archivos diferentes
for chain in structure.get_chains():
    io.set_structure(chain)
    io.save("5AUX_" + chain.get_id() + ".pdb")

# Asignamos nuestro método py3Dmol como una vista
view=py3Dmol.view()
# Las siguiente líneas se utiliza para agregar la clase addModel
# Esta clase nos permite leer los archivos PDB de nuestra única cadena A
view.addModel(open('5AUX.pdb', 'r').read(),'pdb')
#Aquí leemos primero nuestro archivo de DAPK1 junto Kaempferol.
view.addModel(open('compuesto_kaempferol.sdf', 'r').read(),'sdf')
# Ampliación de todas las estructuras visualizadas
view.zoomTo()
# Color de fondo
view.setBackgroundColor('white')
# Estilo de visualización de nuestra cadena A
view.setStyle({"stick": {}})
#view.setStyle({'ligand_pdb'},{'stick': {'color':'red'}})
view.setStyle({'chain':'A'},{'cartoon': {'color':'purple'}})
# Visualización de estructura
view.show()
```
